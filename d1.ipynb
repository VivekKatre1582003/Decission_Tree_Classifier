```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
```

```python
x_train,x_test,y_train,y_test = train_test_split(iris.data,iris.target,random_state=1)
```

```python
x_train
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
pip install pydotplus
```

```python
import pydotplus
from sklearn.tree import export_graphviz
dot_data = export_graphviz(clf,out_file=None)
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_pdf("iris.pdf")
display(graph)

import graphviz
from sklearn.tree import export_graphviz
from IPython.display import display, SVG

# Assuming clf is already defined and trained
# Replace clf with your decision tree classifier object

# Generate DOT data
dot_data = export_graphviz(clf, out_file=None, feature_names = iris.feature_names, class_names = iris.target_names) #This added feature names

# Create graph from DOT data
graph = graphviz.Source(dot_data)

# Save graph as SVG file
graph.render("iris", format="png")

# Display the decision tree as SVG
display(SVG(graph.pipe(format='svg')))



#You can do so much using graphviz... for example you can change boxes .. formatting like things
```

```python
y_train_pred = clf.predict(x_train)
y_test_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train,y_train_pred)
```

```python
confusion_matrix(y_test,y_test_pred)
```
